```python
def convert_html_into_dict(arr, driver):
    
    articles = []
    for article in arr:
        driver.execute_script("arguments[0].scrollIntoView();", article)
        title = article.find_element_by_css_selector("h2").text
        link = article.get_attribute('href')
        img_url = article.find_element_by_css_selector("img").get_attribute("src")
        articles.append({'title': title, 'link': link, 'img_url': img_url})
    
    return articles
```

```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome("./chromedriver")
driver.get("http://tvn24.pl")
driver.implicitly_wait(5)
rodo = driver.find_element_by_css_selector("#rodoLayer > div > div.rodoFooter > a.rodoFooterBtnAccept")
rodo.click()
articles_html = driver.find_elements_by_css_selector("article.default-teaser--standard .link__content > a")

articles = convert_html_into_dict(articles_html, driver)
driver.close()
```

```python
print(articles)
```

```python
def generate_html_str(arr):
    html_str = ""
    for article in arr:
        link = article["link"]
        img_url = article["img_url"]
        html_str += f"""<section style=' width: 500px; background:lightblue'><a href='{link}' style='text-decoration: none'>
        <h2 style='color: black'>{article['title']}</h2>
        <img src='{img_url}'/>
        </a>
        </section>"""
    return html_str
```

```python
display(HTML(generate_html_str(articles)))
```
