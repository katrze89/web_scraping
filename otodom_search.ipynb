```python
import scrapy
```

```python
from scrapy.crawler import CrawlerProcess
from bs4 import BeautifulSoup
```

```python
import logging
from scrapy.utils.log import configure_logging
```

```python
configure_logging(install_root_handler=False)
logging.basicConfig(
filename="otodomlog.txt",
    format="%(levelname)s: %(message)s",
    level= logging.DEBUG
)
```

```python
pages_all = 10
real_estates = []

class OtoDomBot(scrapy.Spider):
    name = "otodom_bot"
    
    def start_requests(self):
        base = "https://www.otodom.pl/sprzedaz/krakow/"
        
        for page_number in range(1, pages_all + 1):
            url = f"{base}?page={page_number}"
            yield scrapy.Request(url=url, callback=self.parse)
    
    def parse(self, response):
        urls = response.css(".offer-item-details > header > h3 > a::attr(href)").extract()
        
        for url in urls:
            self.logger.warning(url)
            yield scrapy.Request(url=url, callback=self.parse_one_page)
    
    def parse_one_page(self, response):
        price = response.css("#root header>div:last-child>div>div:last-child::text").extract_first()
        self.logger.warning(price)
        if price is not None:
            record = {
                'price': price,
                'url': response.request.url
            }
            real_estates.append(record)
```

```python
process = CrawlerProcess()
```

```python
process.crawl(OtoDomBot)
process.start()
```
