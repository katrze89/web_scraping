```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
with urlopen("http://tvn24.pl/") as url:
    b = BeautifulSoup(url.read(), 'html.parser')
    titles_html = b.select("article.default-teaser--standard .link__content > a")
    titles = [title.text for title in titles_html]
    print(titles_html[0])
    print(len(titles))
```

```python
def convert_html_into_dict(arr):
    
    articles = []
    for article in arr:
        title = article.select("h2")[0].text
        link = article.get('href')
        articles.append({'title': title, 'link': link})
    
    return articles
```

```python
convert_html_into_dict(titles_html)
```
